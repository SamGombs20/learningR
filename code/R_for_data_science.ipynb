**DAY 6**
 Data transformation:- row manipulation using arrange, distinct, filter while column manipulation using select, relocate, rename and mutate

```r
library(nycflights13)
```

```r
head(flights)
```

```r
flights |> filter(month==1)
```

```r
glimpse(flights)
```

```r
flights |> arrange(desc(dep_delay))
```

```r
flights |> distinct(origin, dest, .keep_all = TRUE) |> glimpse()
```

```r
flights |> mutate(gain= dep_delay- arr_delay) |> glimpse()
```

```r
x=2
case_when(
    x==1 ~ "You're on!",
    .default = "Default")
```

```r
flights |> mutate(speed = distance / air_time *60)|>glimpse()
```

```r
flights |> select(year, month, day) |> glimpse()
```

```r
flights |> rename(tail_num = tailnum)|> glimpse()
```

```r
flights |> filter(dest=="IAH")|> mutate(speed = distance / air_time * 60)|>select(year, month, day, dep_time, carrier, speed)|> arrange(desc(speed))|>head()
```

```r
flights|>group_by(month)|>glimpse()
```

```r
flights |> group_by(month)|>summarize(avg_delay = mean(dep_delay, na.rm=TRUE), n=n())
```
